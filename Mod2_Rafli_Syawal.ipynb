```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Display top 7 rows of the dataframe in your notebook

```python
dataset = pd.read_csv('nyc-rolling-sales.csv')
dataset.head(7)
```

## Drop column it is empty

```python
dataset.info()
```

```python
dataset = dataset.drop(columns = ['EASE-MENT']).reset_index(drop=True)
dataset
```

```python
dataset = dataset.drop(columns = ['Unnamed: 0']).reset_index(drop=True)
dataset
```

## Check duplicates entri

```python
sum(dataset.duplicated(dataset.columns))
```

```python
dataset = dataset.drop_duplicates(dataset.columns, keep = 'last')
sum(dataset.duplicated(dataset.columns))
```

```python
dataset.info()
```

## Convert data type 

```python
dataset['TAX CLASS AT TIME OF SALE'] = dataset['TAX CLASS AT TIME OF SALE'].astype('category')
dataset['TAX CLASS AT PRESENT'] = dataset['TAX CLASS AT PRESENT'].astype('category')
dataset['LAND SQUARE FEET'] = pd.to_numeric(dataset['LAND SQUARE FEET'], errors='coerce')
dataset['GROSS SQUARE FEET']= pd.to_numeric(dataset['GROSS SQUARE FEET'], errors='coerce')
dataset['SALE DATE'] = pd.to_datetime(dataset['SALE DATE'], errors='coerce')
dataset['SALE PRICE'] = pd.to_numeric(dataset['SALE PRICE'], errors='coerce')
dataset['BOROUGH'] = dataset['BOROUGH'].astype('category')
```

```python
dataset.info()
```

## Checking Null Value

```python
dataset.isnull().sum()
```

```python
dataset.columns[dataset.isnull().any()]
```

```python
data_missing_value = dataset.isnull().sum().reset_index()
data_missing_value.columns = ['feature', 'missing_value']
data_missing_value['percentage'] = round((data_missing_value['missing_value']/len(dataset)) * 100,2)
data_missing_value = data_missing_value.sort_values(['percentage'], ascending=False).reset_index(drop=True)
data_missing_value = data_missing_value[data_missing_value['percentage'] > 0].reset_index(drop=True)
data_missing_value
```

```python
fig, ax = plt.subplots(figsize = (15,7))

g = sns.barplot(x = 'feature', y = 'percentage', data = data_missing_value, ax = ax, 
               palette = sns.color_palette('Blues_d', n_colors=13, desat = 1))
x = np.arange(len(data_missing_value['feature']))
y = data_missing_value['percentage']

for i, v in enumerate (y):
    ax.text(x[i] - 0.1, v + 3, str(v) + '%', fontsize = 20, color = 'gray', fontweight = 'bold')

Title = """
Distribusi Nilai yang Hilang
"""

ax.text(0.8, 80, Title, horizontalalignment = 'right', color = 'black', fontsize = 22, fontweight = 'bold')

text = """
Terdapat 3 fitur yang memiliki missing value (nilai yang hilang), 
3 data numerik
Gross Square Feet memiliki 32.69% missing value,
Land Square Feet memiliki 31.10% missing value,
Sale Price memiliki 16.92% missing value
"""

ax.text(-0.2, 60, text, horizontalalignment = 'left', color = 'black', fontsize = 16, fontweight = 'normal')

ax.set_ylim(0,100)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 0)
plt.tight_layout()
```

## Plot two of the numerical columns one against the other in a scatter plot

```python
features = ['SALE PRICE','TOTAL UNITS','YEAR BUILT','LAND SQUARE FEET','GROSS SQUARE FEET']
sns.pairplot(dataset_numeric[features], 
             diag_kind = 'kde',
             plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k', 'color':'green'},
             size = 4);

plt.savefig('numerical pairplot.png')
```

## Plot the data distribution of all your numerical columns in histogram and boxplot

```python
features = ['SALE PRICE','TOTAL UNITS']
fig, ax = plt.subplots(2,2, figsize=(23,25))

for i in range(len(features)):
    g = sns.distplot(np.log1p(dataset_numeric[features[i]]+1),kde=True, ax=ax[i,0])
    ax[i,0].set_title(features[i])
    ax[i,0].set_xlabel('')
    
    g = sns.boxplot(np.log1p(dataset_numeric[features[i]]+1), color='green', orient='h', ax=ax[i,1])
    ax[i,1].set_title(features[i])
    ax[i,1].set_xlabel('')
```
